```python
def plot_bars(timings, titles, min_val=0, max_val=5):
    plt.rcParams["figure.figsize"] = (20,5)
    np.set_printoptions(precision=1)

    fig, ax = plt.subplots(1, 4, sharey=True)

    data = {'apples': 10, 'oranges': 15, 'lemons': 5, 'limes': 20}
    names = list(data.keys())
    values = list(data.values())

    ax[0].set_ylabel("Execution time (sec)")

    for idx, timing_file in enumerate(timings):    

        timing = np.loadtxt(timing_file, delimiter=",")[:, 1]

        ax[idx].bar(np.arange(len(timing)), timing)

        ax[idx].set_xlabel("Episode count")
        ax[idx].set_ylim((min_val, max_val))

        ax[idx].set_title(f"{titles[idx]} : Mean {np.mean(timing):.2f}")
    plt.show()
```

```python
timing1 = "timing_half_capacity.csv"
timing2 = "timing_leave_previous.csv"
timing3 = "timing_reset_to_init_problem.csv"
timing4 = "timing_zero_each_time.csv"

timings = [
    timing1, 
    timing2,
    timing3,
    timing4
]

titles = [
    "Half capacity Init",
    "Previous flow Init",
    "'Do nothing' problem Init",
    "Zero Init"
]

plot_bars(timings, titles, min_val=3.5, max_val=5.5)
```

```python
timing1 = "timing_beta_2.csv"
timing2 = "timing_beta_3.csv"
timing3 = "timing_beta_4.csv"

timings_beta = [
    timing1, 
    timing2,
    timing3,
]

titles_beta = [
    "Beta = 2",
    "Beta = 3",
    "Beta = 4"
]

plot_bars(timings_beta, titles_beta, min_val=0.5, max_val=5.5)
```
