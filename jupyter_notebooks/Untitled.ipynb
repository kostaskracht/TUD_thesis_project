```python
# Plot the costs
def smooth(rewards, wnd, save=None):
    rew_smooth = []
    for i in range(wnd, len(rewards)):
        rew_smooth.append(-np.mean(rewards[i:i+wnd]))
    return rew_smooth

# Plot the rewards over iterations
def plot_rewards(rewards, benchmarks = {}, smooth_window=None, save=True, cum=False):
    
    fig, ax = plt.subplots()
    
    norm_factors = [9.53120000e+01, 5.88369327e+08, 3.05471214e+08] 
    labels = ["Lifecycle cost", "CO2e emissions", "Total travel time"]
    for idx, reward in enumerate(rewards):
        reward = reward/np.sum(reward)
#         import pdb; pdb.set_trace()
        if cum:
            reward = np.cumsum(reward)

        if smooth_window:
            reward = smooth(reward, smooth_window)
    #     plt.yscale('log')
        
        ax.plot(reward, label=labels[idx])
#         colors = {"time": "r", "condition": "g"}
#         for key,value in benchmarks.items():
#             plt.axhline(y=value, color=colors[key], linestyle='-', label=key)
    #         plt.axhline(y=137, color='r', linestyle='-')
        
        
        

        if save: 
            pass
    #         upload_to_html(fig, filename)
    plt.xlabel("episodes")
    plt.ylabel("lifecycle cost")
    plt.legend()
    plt.show()
    
plot_rewards(costs.T, cum=True)
```
